Handling NaN values

1. Handling NaN values in the director column

```python
print(f'Shape of the column: {df["director"].shape}')
print(f'Number of NaN values: {df["director"].isnull().sum()}')
print(f'Percentage of NaN values: {df["director"].isna().mean() * 100}')
df[df["director"].isnull()]
```

```python
print(df["director"].value_counts())
print("_" * 60)
print(df["director"].value_counts(normalize =True) * 100)
```

```python
# There are around 30% of missing director names
# Encoding only high frequency directors and others as "other"

top_directors = df["director"].value_counts().nlargest(10).index
df["director"] = df["director"].fillna('Unknown')
df["director"] = df["director"].apply(lambda x: x if x in top_directors else 'Other')
```

```python
df["director"].isnull().sum()
```

2. Handling NaN values in the "Cast" column

```python
print(f'Shape of the column: {df["cast"].shape}')
print(f'Number of NaN values: {df["cast"].isnull().sum()}')
print(f'Percentage of NaN values: {df["cast"].isna().mean() * 100}')
df[df["cast"].isnull()]
```

```python
print(df["cast"].value_counts())
print("_ " * 60)
print(df["cast"].value_counts(normalize = True) * 100)
```

```python
# 9% values are missing
# Replacing them with "Unknown"

df["cast"] = df["cast"].fillna('UnKnown')
```

```python
df["cast"].isnull().sum()
```

3. Handling NaN values in the country column

```python
print(f'Shape of the column; {df["country"].shape}')
print(f'Number of NaN values: {df["country"].isnull().sum()}')
print(f'Percentage of NaN values: {df["country"].isnull().mean() * 100}')
df[df["country"].isnull()]
```

```python
print(df["country"].value_counts())
print("-" * 60)
print(df["country"].value_counts(normalize = True) * 100)
```

```python
# Using the mode of the column

df["country"].fillna(df["country"].mode()[0], inplace = True)
```

```python
df["country"].isnull().sum()
```

4. Handling NaN values in the date_added column

```python
print(f'Shape of the Column: {df["date_added"].shape}')
print(f'Number of NaN values: {df["date_added"].isnull().sum()}')
print(f'Percentage of NaN values: {df["date_added"].isnull().mean() * 100}')
df[df["date_added"].isnull()]
```

```python
print(df["date_added"].value_counts())
print("-" * 60)
print(df["date_added"].value_counts(normalize = True) * 100)
```

```python
# Only 10 values are missing, so replacing them with the mode of the column

df["date_added"].fillna(df["date_added"].value_counts().idxmax(), inplace = True)
```

```python
df["date_added"].isnull().sum()
```

5. Handling NaN values in the rating column

```python
print(f'Shape of the column: {df["rating"].shape}')
print(f'Number of NaN values: {df["rating"].isnull().sum()}')
print(f'Percentage of NaN values: {df["rating"].isnull().mean() * 100}')
df[df["rating"].isnull()]
```

```python
print(df["rating"].value_counts())
print("-" * 60)
print(df["rating"].value_counts(normalize = True) * 100)
```

```python
# Replacing with the mode value of the column
df["rating"].fillna(df["rating"].value_counts().idxmax(), inplace = True)
```

```python
df["rating"].isnull().sum()
```

```python
print(f'Shape of the column: {df["duration"].shape}')
print(f'Number of NaN values: {df["duration"].isnull().sum()}')
print(f'Percentage of NaN values: {df["duration"].isnull().mean() * 100}')
df[df["duration"].isnull()]
```
